```python
import json
from plot import *
```

```python
# 从outline_cache中读取json
with open("../outline_cache/new_outline.json", "r", encoding='utf-8') as f:
    outline = json.load(f)
```

```python
plot_chains = outline['plot_chains']
```

```python
# 允许将plot_chains中的任何一个分支作为original plot chain
plot_nodes = plot_chains_to_nodes_tree(plot_chains, root_info='开拓')
main_character = list(outline['characters'].keys())[0]
draw_node_tree(plot_nodes[0], save_name='main')
```

```python
# 保存为json，index作为key
# 子节点也改存对应的数字编号
game_plot_nodes_dict = {
    "characters": outline['characters'],
    "initial_relationship": outline['initial_relationship'],
}

# 先顺序的存子节点信息
def dfs_convert(node):
    children_index = []
    for child in node.children:
        children_index.append(plot_nodes.index(child))
    game_plot_nodes_dict[plot_nodes.index(node)] = {
        'node_info': node.node_info,
        'summary': node.summary,
        'start': node.start,
        'end': node.end,
        'characters': node.characters_behavior,
        'location': node.location,
        'time': node.time,
        'children_index': children_index,
    }
    
    for child in node.children:
        dfs_convert(child)

dfs_convert(plot_nodes[0])

with open("../outline_cache/game_plot_nodes.json", "w", encoding='utf-8') as f:
    json.dump(game_plot_nodes_dict, f, ensure_ascii=False, indent=4)
```
